```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv("C:\\Users\\Amir\\OneDrive\\Desktop\\sales_data.CSV")
```

```python
sales.drop_duplicates(subset=None, keep=False)
```

```python
sns.countplot(x='Customer_Gender', data=sales)
plt.title('Distribution of Customer Genders')
plt.xticks(rotation=45)

total = float(len(sales))
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2, height + 3, '{:.1f}%'.format((height / total) * 100), ha='center', fontsize=12)
```

```python
plt.hist('Customer_Age', data=sales[sales['Customer_Gender'] == 'M'], label= 'Male')
plt.hist('Customer_Age', data=sales[sales['Customer_Gender'] == 'F'], label= 'Female')
plt.title('Distribution of Age by Gender')
plt.xlabel('Age')
plt.legend()
```

```python
sns.countplot(x='Age_Group', data=sales)
plt.title('Distribution of Age Groups')
plt.xticks(rotation=45)

total = float(len(sales))
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2, height + 3, '{:.1f}%'.format((height / total) * 100), ha='center', fontsize=12)
```

```python
labels= ['Accessories', 'Bikes', 'Clothing']

colors=['blue', 'yellow', 'red']

sizes= [1500, 600, 500]

plt.pie(sizes,labels=labels, colors=colors, startangle=90, autopct='%1.2f%%')

plt.title('Product_Category')

plt.axis('equal')

plt.show()
```
